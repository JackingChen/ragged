```python
import json
input_file = "/data/user_data/jhsia2/dbqa/data/kilt_knowledgesource.json"
output_file = "/data/user_data/jhsia2/dbqa/data/id2title.json"
num_par_docs = 0
id2title = dict()
with open(input_file, 'r') as infile:
    for l_id, line in enumerate(infile):
        if (l_id%100000==0):
            print(l_id)
        data = json.loads(line)
        id2title[data['wikipedia_id']] = data['wikipedia_title']
print('done')
print('num_par_docs = ', num_par_docs)
```

```python
with open('/data/user_data/jhsia2/dbqa/data/id2title.json', 'w') as file:
    # Use json.dump to write the list of dictionaries to the file
    json.dump(id2title, file, indent=4)
```
