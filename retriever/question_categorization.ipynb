```python
import json

def load_jsonl(file_path):
    print('loading from', file_path)
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            json_obj = json.loads(line.strip())
            # break
            data.append(json_obj)
    return data
```

```python
data = load_jsonl('/data/user_data/jhsia2/dbqa/data/nq-dev-kilt.jsonl')
```

```python
types = ['who', 'what', 'when', 'where', 'why']
type_count = {}
for t in types:
    type_count[t] = []
type_count['other'] = []
for d in data:
    first_word = d['input'].strip().split()[0]
    if first_word in types:
        type_count[first_word].append({'id': d['id'], 'input': d['input']})
    else:
        type_count['other'].append({'id': d['id'], 'input': d['input']})
```

```python
for t in type_count:
    print(t, len(type_count[t]))
```

```python
import re

# questions = ["How many apples are there?", "Who is the president?", "When was the Eiffel Tower built?", "How does a computer work?"]

# Patterns for different question types
who_pattern = r"^\s*who\b"
what_pattern = r"^\s*(what\b|which\b)"
when_pattern = r"^\s*when\b"
where_pattern = r"^\s*(where\b)"
why_pattern = r"^\s*why\b"
how_process_pattern = r"^\s*how\b(?! many\b| much\b| often\b| long\b| far\b)"
numerical_pattern = r"^\s*(how many\b|how much\b|how often\b|how long\b|how far\b)"

# Dictionary to store categorized questions
categorized_questions = {
    "Who": [],
    "What": [],
    "When": [],
    "Where": [],
    "Why": [],
    "How (Process)": [],
    "Numerical": [], 
    "Other": []
}

# Categorizing questions
for d in data:
    index = d['id']
    question = d['input']
    if re.match(who_pattern, question, re.IGNORECASE):
        categorized_questions["Who"].append((index, question))
    elif re.match(what_pattern, question, re.IGNORECASE):
        categorized_questions["What"].append((index, question))
    elif re.match(when_pattern, question, re.IGNORECASE):
        categorized_questions["When"].append((index, question))
    elif re.match(where_pattern, question, re.IGNORECASE):
        categorized_questions["Where"].append((index, question))
    elif re.match(why_pattern, question, re.IGNORECASE):
        categorized_questions["Why"].append((index, question))
    elif re.match(how_process_pattern, question, re.IGNORECASE):
        categorized_questions["How (Process)"].append((index, question))
    elif re.match(numerical_pattern, question, re.IGNORECASE):
        categorized_questions["Numerical"].append((index, question))
    else:
        categorized_questions["Other"].append((index, question))

# Display the categorized questions
for category, questions in categorized_questions.items():
    print(f"{category} Questions: {len(questions)}")
```

```python
import re

# Sample list of questions
# questions = [
#     "What is the capital of France?",
#     "Who wrote 'To Kill a Mockingbird'?",
#     "How many continents are there on Earth?",
#     "Is the sky blue?",
#     "Explain the theory of relativity.",
#     "List the ingredients in a Caesar salad."
# ]

# Regex patterns for different question types
# r"^\s*(what|who|where|when|which)\b|\bis called$"
factoid_pattern = r"^\s*(what|who|where|when|which|on which|how long)\b|\bis called$"
yes_no_pattern = r"^\s*(is|did)\b"
numerical_pattern = r"^\s*how many\b"
explanation_pattern = r"^\s*explain\b"
list_pattern = r"^\s*list\b"

# Function to categorize question
def categorize_question(question):
    if re.match(factoid_pattern, question, re.IGNORECASE):
        return "Factoid"
    elif re.match(yes_no_pattern, question, re.IGNORECASE):
        return "Yes/No"
    elif re.match(numerical_pattern, question, re.IGNORECASE):
        return "Numerical"
    elif re.match(explanation_pattern, question, re.IGNORECASE):
        return "Explanation"
    elif re.match(list_pattern, question, re.IGNORECASE):
        return "List"
    else:
        return "Unknown"

# Dictionary to store categorized questions
categorized_questions = {
    "Factoid": [],
    "Yes/No": [],
    "Numerical": [],
    "Explanation": [],
    "List": [],
    "Unknown": []
}

# Categorizing each question and adding index
for d in data:
    index = d['id']
    question = d['input']
    category = categorize_question(question)
    categorized_questions[category].append((index, question))

# Display the categorized questions
for category, questions in categorized_questions.items():
    print(f"{category} Questions: {len(questions)}\n")
```
